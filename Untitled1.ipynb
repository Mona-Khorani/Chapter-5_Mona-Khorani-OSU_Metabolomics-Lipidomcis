```python
#HexCer
#there are some issue with - and pos values, i put intensities that is high as neg value and vise 


import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


#Hip FCI/FCS
stacked_array = np.array([1.9251, 0.39387, -0.98652,8.4179,-1.1457,-1.0014,-0.16611,2.4561])
x_positions = [32, 36, 36, 40,  42, 42, 42, 44]
y_positions = [1, 1, 2, 1,  0, 1, 2, 2]
#-8.4179, 0.9253, 1.5392, 1.3175
#0.16611,-0.024611



#Cortx NOI/NOS
#stacked_array = np.array([-0.49246, -0.081629, 0.64585,-0.37657, 0.96132,0.56044,- 0.79281,-0.3435])
#x_positions = [32, 36, 36, 40,  42, 42, 42, 44]
#y_positions = [1, 1, 2, 1,  0, 1, 2, 2]
#0.56044, -1.8087, 0.37657, 
#0.79281


#Cortx FCI/FCS
#stacked_array = np.array([0.2022, +0.27994, -1.0102,0.87512, +0.676, -1.5912, +1.9167, -0.020534])
#x_positions = [32, 36, 36, 40,  42, 42, 42, 44]
#y_positions = [1, 1, 2, 1,  0, 1, 2, 2]

# Convert x_positions and y_positions to numpy arrays
x_positions = np.array(x_positions) - 30  # Shift x positions by subtracting 30
y_positions = np.array(y_positions)

# Determine the dimensions of the heatmap
num_rows = max(y_positions) + 1  # Add 1 to include the last row
num_cols = max(x_positions) + 1  # Add 1 to include the last column

# Create an empty grid for the heatmap
heatmap = np.zeros((num_rows, num_cols))
heatmap.fill(np.nan)

# Assign the intensity values to their corresponding positions in the grid
for i in range(len(stacked_array)):
    x_pos = x_positions[i]
    y_pos = y_positions[i]
    intensity = stacked_array[i]
    heatmap[y_pos, x_pos] = intensity

# Create a custom colormap for intensity from -1.5 to 1.5
colors = ['#67001f', '#b2182b', '#d6604d', '#f4a582', '#fddbc7', 'white', '#d1e5f0', '#92c5de', '#4393c3', '#2166ac', '#053061']
cmap_custom = mcolors.LinearSegmentedColormap.from_list('custom_cmap', colors, N=256)

# Set the color for NaN values to gray
cmap_custom.set_bad(color='gray')

# Set the color boundaries for the color map
vmin =- 9
vmax = 9
norm = mcolors.TwoSlopeNorm(vmin=vmin, vcenter=0, vmax=vmax)
bounds = np.linspace(vmax, vmin, 11)
cmap_custom.set_over('white')
cmap_custom.set_under('white')  # Set the lowest color to blue

# Create the figure and axes with square aspect ratio
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_aspect('equal')  # Set aspect ratio to make boxes square

# Set the background color of the plot area to dark gray
ax.set_facecolor('darkgray')

# Create the heatmap using pcolormesh with the custom colormap
im = ax.pcolormesh(heatmap, cmap=cmap_custom, edgecolor='darkgray', linewidth=0.4, norm=norm)

# Add x and y labels
ax.set_xlabel('Number of fatty-acid carbon atoms', labelpad=10)
ax.set_ylabel('Number of fatty-acid double bonds', labelpad=10)

# Adjust the grid line positions for size 1x1
x_ticks = np.arange(num_cols) + 0.5
y_ticks = np.arange(num_rows) + 0.5

# Add grid lines for each box with size 1x1
ax.set_xticks(x_ticks, minor=False)
ax.set_yticks(y_ticks, minor=False)
ax.grid(True, color='darkgray', linewidth=0.1)

# Adjust the x-axis tick positions and labels
x_tick_positions = np.arange(num_cols)
x_tick_labels = np.arange(num_cols) + 30
ax.set_xticks(x_tick_positions + 0.5)
ax.set_xticklabels(x_tick_labels)

# Adjust the y-axis tick positions and labels
y_tick_positions = np.arange(num_rows)
y_tick_labels = np.arange(num_rows)
ax.set_yticks(y_tick_positions + 0.5)
ax.set_yticklabels(y_tick_labels)

# Create the color bar with adjusted size and reversed colors
cbar = fig.colorbar(im, ax=ax, fraction=0.035, pad=0.04, ticks=[vmin, 0, vmax])
cbar.set_ticklabels(['{:.1f}'.format(vmin), '0', '{:.1f}'.format(vmax)])
cbar.ax.invert_yaxis()  # Reverse the color bar

# Save the plot with a dpi of 400
plt.savefig('heatmap.png', dpi=400)

# Display the heatmap
plt.show()
```
